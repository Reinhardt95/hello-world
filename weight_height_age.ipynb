```python
import numpy as np; import matplotlib.pyplot as plt

#weight measured in pounds (pounds, standard deviation, sample size)
weight = np.round(np.random.normal(172, 29, 100), 2)
#height measured in inches (inches, standard deviation, sample size)
height = np.round(np.random.normal(69, 2.9, 100), 2)
age = np.round(np.random.normal(35, 4.5, 100), 2)

np_weight = np.array(weight)
np_height = np.array(height)
np_age = np.array(age)

#sample of 100 males from a city)
np_city =  np.column_stack((weight, height, age))

print(np_city)
```

```python
plt.scatter(weight, height)
plt.show()
```

```python
plt.xlabel("Weight in pounds")
plt.ylabel("Height in inches")
plt.title("Correlation between weight and heigth in 100 male sample")

plt.scatter(x = np_weight, y = np_height, s = np.array(age) * 2, alpha = 0.7)
plt.show()
```

```python
plt.hist(np_weight, np_height, np_age)
plt.show()
```

```python
plt.hist(np_weight, np_height)
plt.show()
```

```python
plt.hist(np_weight, bins=5, width=5.0, alpha=0.8)
plt.grid(True)
plt.show()
```

```python
print(np.corrcoef(np_weight, np_height))
```

```python
print(np.corrcoef([np_weight, np_height, np_age]))
```
